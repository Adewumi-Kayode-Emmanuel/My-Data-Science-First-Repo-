```python
# Initial account balance
balance = 1000.0

# Function to display the current balance
def check_balance(current_balance):
    print(f"Your current balance is: ₦{current_balance:.2f}")

# Function to handle deposits
def deposit(current_balance):
    try:
        amount = float(input("Enter amount to deposit: ₦"))
        if amount > 0:
            current_balance += amount
            print(f"₦{amount:.2f} deposited successfully.")
        else:
            print("Deposit amount must be positive.")
    except ValueError:
        print("Invalid input. Please enter a valid number.")
    return current_balance

# Function to handle withdrawals
def withdraw(current_balance):
    try:
        amount = float(input("Enter amount to withdraw: ₦"))
        if amount > 0:
            if amount <= current_balance:
                current_balance -= amount
                print(f"₦{amount:.2f} withdrawn successfully.")
            else:
                print("Insufficient funds.")
        else:
            print("Withdrawal amount must be positive.")
    except ValueError:
        print("Invalid input. Please enter a valid number.")
    return current_balance

# Main banking system function
def banking_system():
    global balance # Use the global balance variable

    while True:
        print("\n--- Welcome to your bank account! ---")
        print("1. Check Balance")
        print("2. Deposit")
        print("3. Withdraw")
        print("4. Exit")

        choice = input("Choose an option: ")

        if choice == '1':
            check_balance(balance)
        elif choice == '2':
            balance = deposit(balance)
        elif choice == '3':
            balance = withdraw(balance)
        elif choice == '4':
            print("Thank you for banking with us. Goodbye!")
            break
        else:
            print("Invalid option. Please choose 1, 2, 3, or 4.")

# Start the banking system
banking_system()
```
